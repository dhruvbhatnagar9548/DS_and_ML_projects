# Thresholding

It is a technique in which every pixel value of image is compared 
with the provided threshold value and the required operation applied according to provided techniques. It is a very popular segmentation technique for sepreating the object from foreground to background. This applied on the greyscale images therefore images need to converted to greyscale.


#### There are diffrent type of basic thresholding techniques:
1. Binary- If pixel is greater than set threshold value set to 255(white) else 0(black) 
2. Binary_inv- just opposite of Binary
3. TOZERO- pixel will be set to 0 if it is less than provided threshold value
4. TOZERO_inv- just opposite of TOZERO
5. TRUNC- If pixel greater than threshold converted to set value else remains same

```python
image_threshold = cv2.cvtColor(image_cvt, cv2.COLOR_BGR2GRAY)
fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(image_threshold)
plt.title("Original_image")

# this returns 2 outputs. First is the threshold that is used and second threshold image
# cv2.THRESH_BINARY
fig.add_subplot(rows, columns, 2)
ret, thresh_1 = cv2.threshold(image_threshold, 150, 225, cv2.THRESH_BINARY)
plt.imshow(thresh_1)
plt.title('Binary_threshold_image')

# cv2.THRESH_BINARY_INV
fig.add_subplot(rows, columns, 3)
ret, thresh2 = cv2.threshold(image_threshold, 150, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh2)
plt.title('Binary_Inverse_threshold_image')

# cv2.THRESH_TOZERO
fig.add_subplot(rows, columns, 4)
ret, thresh3 = cv2.threshold(image_threshold, 150, 255, cv2.THRESH_TOZERO)
plt.imshow(thresh3)
plt.title('Tozero_threshold_image')

# cv2.THRESH_TOZERO_INV
fig.add_subplot(rows, columns, 5)
ret, thresh4 = cv2.threshold(image_threshold, 150, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thresh4)
plt.title('Tozero_inverse_threshold_image')

# cv2.THRESH_TRUNC
fig. add_subplot(rows, columns, 6)
ret, thresh5 = cv2.threshold(image_threshold, 150, 255, cv2.THRESH_TRUNC)
plt.imshow(thresh5)
plt.title('Truncated_image')

plt.show()
```

##### Adaptive Thresholding

In a basic thresholding global value remains constant. However if lightning conditions are not good than constant threshold is not useful. There is the requiremnt of variable change in the constant depending on the lightning conditions. Therefore threshold values changes in a smaller region.

So there are two Adaptive threshold techniques measurely used are:
1. Adaptive Threshold MEAN- threshold value is a mean of the neighbourhood area minus the constant
2. Adaptive Threshold GAUSSIAN- threshold value is a gaussian-weighted sum of the neighbourhood values minus the constant

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
column = 3
fig.add_subplot(rows,column,1)
image_adaptive_threshold_original = cv2.imread("bp1.jpg")
image_adaptive_threshold = cv2.cvtColor(image_adaptive_threshold_original, cv2.COLOR_BGR2GRAY)
plt.imshow(image_adaptive_threshold)
plt.title("Original image")

# cv2.ADAPTIVE_THRESH_MEAN_C
fig.add_subplot(rows,column,2)
th_mean = cv2.adaptiveThreshold(image_adaptive_threshold,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,181,5)
plt.imshow(th_mean)
plt.title('Adaptive_threshold_MEAN')

# cv2.ADAPTIVE_THRESH_GAUSSIAN_C
fig.add_subplot(rows,column,3)
th_gaussian = cv2.adaptiveThreshold(image_adaptive_threshold,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,181,5)
plt.imshow(th_gaussian)
plt.title('Adaptive_threshold_GAUSSIAN')
plt.show()
```

### Bitwise Operation
Bitwise operations can be used for extracting and manipulation of the image.

1. AND- This condition will be TRUE if both pixels are greater than 0
2. OR- This condition will be TRUE if any image of both pixels are greater than 0
3. XOR- It is TRUE only if any of the pixel value greater tha 0 but not both should be greater than 0
4. NOT- It will be inverse if 0-> 255 or 255->0

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
column = 2
fig.add_subplot(rows,column,1)
image_circle = cv2.imread('circle.jfif')
plt.imshow(image_circle)
print('Image circle shape- ',(image_circle.shape))

fig.add_subplot(row,column,2)
image_rectangle = cv2.imread('rectangl.png')
image_rectangle = cv2.resize(image_rectangle, dsize=(254,198))
print('Image rectangle shape- ',(image_rectangle.shape))
plt.imshow(image_rectangle)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 2
column = 2

# Bitwise_AND
fig.add_subplot(rows,column,1)
image_AND = cv2.bitwise_and(image_circle, image_rectangle)
plt.imshow(image_AND)
plt.title("Image_AND")

# Bitwise OR
fig.add_subplot(rows,column,2)
image_OR = cv2.bitwise_or(image_circle, image_rectangle)
plt.imshow(image_OR)
plt.title("Image_OR")

# Bitwise XOR
fig.add_subplot(rows,column,3)
image_XOR = cv2.bitwise_xor(image_circle, image_rectangle)
plt.imshow(image_XOR)
plt.title("Image_XOR")

# Bitwise NOT
fig.add_subplot(rows,column,4)
image_NOT = cv2.bitwise_not(image_circle)
plt.imshow(image_NOT)
plt.title("Image_NOT")
plt.show()
```

### Translating an Image

It means the rectilinear shift of the image from one location to another. To do this the cv2.WrapAffine method. This function requires a 2x3 array. The numpy array should be flot type.

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
column = 2

fig.add_subplot(rows,column,1)
image_translate = image.copy()
plt.imshow(image_translate)
plt.title('Image')

fig.add_subplot(rows,column,2)
print("Shape of the image-", image_translate.shape)
height = image_translate.shape[0]
width = image_translate.shape[1]

image_MATRIX = np.float32([[1,0,100],[0,1,100]])
shifting = cv2.warpAffine(image_translate,image_MATRIX,(width,height))
plt.imshow(shifting)
plt.title('Shifting of the image')
plt.show()
```

## Morphological operation

### Basic morphological operations

1. Erosion- Decrease the noise. sharpening the image. If orignal image let's say made up of (either 1 or 0) will be considered 1 only if all pixels under kernel are 1 else it will become 0.
2. Dilation- Increase the noise. dimnishing an image. Similarly orignal image made up of either 1 or 0 will be considered 1 if one pixel under the kernel is 1.

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
column = 3

fig.add_subplot(rows,column,1)
image_ero_dil = image.copy()
plt.imshow(image_ero_dil)
plt.title('Image')

fig.add_subplot(rows, column, 2)
kernel = np.ones((5,5), np.uint8)
img_erosion = cv2.erode(image_ero_dil, kernel, iterations=5)
plt.imshow(img_erosion)
plt.title('Eroded image')

fig.add_subplot(rows, column, 3)
img_dilate = cv2.dilate(image_ero_dil, kernel, iterations=5)
plt.imshow(img_dilate)
plt.title('Dilated image')
plt.show()
```

##### Morphological operation (continued)

1. Opening- It requires erosion followed by dilation (at foreground) of the image
2. Closing - It requires dilation followed by erosion (at foreground) of the image
3. Morphological gradient - It diffrent from the other operations as mentioned above, in this morphological gradient first applies erosion and dilation individually on the image and then computes the difference between the eroded and dilated image. The output will be an outline of the given image.

```python
fig = plt.figure(figsize=(10, 7))
rows = 2
column = 2

fig.add_subplot(rows,column,1)
image_ero_dil = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_ero_dil)
plt.title("Image")

fig.add_subplot(rows,column,2)
# binarize the image
binr = cv2.threshold(image_ero_dil, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(binr, cv2.MORPH_OPEN, kernel, iterations=5)# opening the image
plt.imshow(opening)
plt.title("Opening Image")

fig.add_subplot(rows,column,3)
closing = cv2.morphologyEx(binr, cv2.MORPH_CLOSE, kernel, iterations=5)# closing the image
plt.imshow(closing)
plt.title("Closing Image")

fig.add_subplot(rows,column,4)
invert = cv2.bitwise_not(binr)
morph_gradient = cv2.morphologyEx(invert,cv2.MORPH_GRADIENT,kernel)# morphological gradient of the image
plt.imshow(morph_gradient)
plt.title("Morphological gradient of the Image")
plt.show()
```
