# Data vizualiation on exchange rates

In this project we will work on the dataset provided by the European central bank which describes Euro daily exchange rate from 1999 - 2020.

Daria Chemkaeva put together the data set and made it available on Kaggle — the data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on January 2021.

##### Goal of the project
We will work on the data to see the how the Euro-US dollar exchange rate changes when COVID pandemic hit to the USA.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.style as style
currency = pd.read_csv("euro-daily-hist_1999_2020.csv")
```

```python
currency.tail()
```

```python
print(currency.shape)
print(currency.info())
# It can be noticed that Dtypes mainly object 
```

```python
currency.columns
```

```python
# changing column name from Camelcase to Snakecase
# Also renaming Period/Unit to Time for understaning quickly
currency.columns = ['Time', 'Australian_dollar', 'Bulgarian_lev',
       'Brazilian_real', 'Canadian_dollar', 'Swiss_franc',
       'Chinese_yuan _renminbi', 'Cypriot_pound', 'Czech_koruna',
       'Danish_krone', 'Estonian_kroon', 'UK_pound_sterling',
       'Greek_drachma', 'Hong_Kong_dollar', 'Croatian_kuna',
       'Hungarian_forint', 'Indonesian_rupiah', 'Israeli_shekel',
       'Indian_rupee', 'Iceland_krona', 'Japanese_yen',
       'Korean_won', 'Lithuanian_litas', 'Latvian_lats',
       'Maltese_lira', 'Mexican_peso', 'Malaysian_ringgit',
       'Norwegian_krone', 'NZ_dollar', 'Philippine_peso',
       'Polish_zloty', 'Romanian_leu', 'Russian_rouble',
       'Swedish_krona', 'Singapore_dollar', 'Slovenian_tolar',
       'Slovak_koruna', 'Thai_baht', 'Turkish_lira', 'US_dollar',
       'South_African_rand']
currency.head()
```

```python
currency["Time"].head()
```

```python
currency["Time"] = pd.to_datetime(currency["Time"])#convert argument to datetime format
currency["Time"].head()
```

```python
currency.sort_values("Time", inplace=True)# sorting the table according to time
currency["Time"].head()# It can be observe that index value got reverse
```

```python
currency.reset_index(drop=True, inplace=True)# reset the index value
currency.head()
```

```python
euro_to_dollar = currency[["Time","US_dollar"]].copy()
euro_to_dollar["US_dollar"].value_counts()
```

```python
# It can be observe that there are 61 records whose value equal to "-". We will only keep numeric values
euro_to_dollar = euro_to_dollar[euro_to_dollar["US_dollar"] != "-"]
euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)
euro_to_dollar.info()
```

```python
euro_to_dollar.head()
```

```python
euro_to_dollar.tail()
```

```python
# Finish cleaning the data I will generate a line plot 
%matplotlib inline
plt.plot(euro_to_dollar["Time"],euro_to_dollar["US_dollar"])
plt.title("Euro-Dollar Exchange rate: Jan, 1999 - Jan, 2021")
plt.xlabel("Year")
plt.ylabel("Euro-Dollar Exchange rate")
plt.show()
```

```python
# It can be observed from the graph that there are many wiggles on the curve it is not smooth 
# This may be due to the daily variation in the exchange rate.
# The rate only shows smooth upwards or downwards trend in the long run such as months or years 
# so long-term can be obtained by rolling mean(moving average)
# As our main goal is to find the effect on the exchange rate on dollar due to CORONA Pandemic
# But first lets find about rolling mean values for 7, 30, 50, 100, 365 days and compare to orignal values

plt.figure(figsize = (9,6))

plt.subplot(2,3,1)
plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.title("Orignal values", weight="bold")

for i,rolling_mean in zip([2,3,4,5,6],[7,30,50,100,365]):
    plt.subplot(2,3,i)
    plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"].rolling(rolling_mean).mean())
    plt.title(f"Rolling Window: {rolling_mean}", weight="bold")
plt.tight_layout()
plt.show()
```

```python
# I will calculate the rolling_mean of 30 days for the US_dollar.
# So, lets make a new column of "rolling_mean".
euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()
print(euro_to_dollar.head(51))
print("The maximum value of rolling mean is {}".format(str(euro_to_dollar["rolling_mean"].max().round(2))))
print("The minimum value of rolling mean is {}".format(str(euro_to_dollar["rolling_mean"].min().round(2))))
```

### Coronavirus impact on US Dollar

As the name suggest the euro-dollar rate has changed during the Coronavirus pandemic. As we know that pandemic arrived in January-2021, we will see the 2020 data(January 2020 to January 2021) and compare to the 2017-2019 baseline.

```python
before_pandemic = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >= 2017) & (euro_to_dollar["Time"].dt.year <=2019)]
during_pandemic = euro_to_dollar.copy()[euro_to_dollar["Time"].dt.year >= 2020]
```

```python
style.use("fivethirtyeight")

# Adding the subplots

fig = plt.figure(figsize=(12,6))
fig.tight_layout()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,1,2)
axes = [ax1,ax2,ax3]

for ax in axes:
    ax.set_ylim(0.8, 1.6)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(["1.0","1.2","1.4", "1.6"], alpha=0.3)
    ax.grid(alpha=0.5)

#before pandemic 2017-2019
ax1.plot(before_pandemic["Time"], before_pandemic["rolling_mean"], color="#2ca02c")
ax1.set_xticklabels(["2017","","", "2018","","", "2019","","", "2020"], fontsize=12,alpha=0.3)
ax1.set_title("EUR-USD Rated (2017-19)", fontsize =15, weight="bold", color="#2ca02c")


ax2.plot(during_pandemic["Time"], during_pandemic["rolling_mean"], color="#d62728")
ax2.set_xticklabels(["Jan\n(2020)","", "April\n(2020)","", "Aug\n(2020))","", "2021"], fontsize=12,alpha=0.3)
ax2.set_title("EUR-USD Rated (2020)", fontsize =15, weight="bold", color="#d62728")



ax3.plot(before_pandemic["Time"], before_pandemic["rolling_mean"],color='#2ca02c')
ax3.plot(during_pandemic["Time"], during_pandemic["rolling_mean"],color='#d62728')
ax3.grid(alpha=0.5)
ax3.set_xticks([])

plt.show()
```

#### Conclusion 
- It can be noticed from the above graph that in the first four month of 2018 the Euro-US dollar exchange rate reaches to its highest point 1.24 than there was a continius decline. But there is a sudden increase in the exchange rate when coronvirus pandemic hit because of the recession.
